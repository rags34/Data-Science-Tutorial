# 1. Pandas Tutorial

## Loading Data into Pandas

```python
import pandas as pd

#Read csv file
df = pd.read_csv('Pokemon.csv')

#Read the last 5 rows in the dataframe
print(df.tail(5)) #print is not really important here in jupyter notebook 
```

## Reading Data 

```python
#Read the name, type1 and type2 columns of first 5 rows in the dataframe
print(df[['Name', 'Type 1', 'Type 2']].head(5)) #Same as df['Name'].head(5),df['Type 1'].head(5),df['Type 2'].head(5)
```

```python
#Read multiple specific Columns
df[['Name','Type 1','Type 2']]
```

```python
#Read a specific Row
df.iloc[1]
```

## iteration through each row

```python
#iterating through the index and rows in the dataset
df2 = df.head(10)

for index, row in df2.iterrows():
    print(index, row)
```

```python
#iterating through the index and specific columns in the dataset
for index, row in df2.iterrows():
    print(index, row[['Name','Type 1', 'Type 2']])
```

## Getting rows based on specific conditons 

```python
# Access a group of rows and columns by label(s) or a boolean array
df.loc[df['Type 1'] =="Fire"] # for bool -> df.loc[df['Legendary']](gives all the data where bool value is True)
```

## High level description of data (min, max, mean, sd, etc.) 

```python
#Gives us all the matrics for numerical values
df.describe()
```

## Sorting data 

```python
#Sorting
df.sort_values('Name', ascending = False)  #Sorts data by Name in descending alphabetical order
```

```python
#Sort values first by 'Type 1'(asc) and then by 'HP'(desc)
df.sort_values(['Type 1', 'HP'] , ascending = [True,False]) #ascending = [1,0] also works here
```

## Adding a column

```python
#adding a column (total of all the numeric stats for each pokemon)
#df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
#Recommended Way
df['Total'] = df.iloc[:, 4:10].sum(axis=1) 
#[:, 4:9] means we take all the rows and from the 4th to the 9th column and then .sum(axis = 1) means adding these horizontally. 
# Similarly, .sum(axis = 0) adds vertically
df.head(5)
```

## Deleting a column 

```python
#Deleting a column 
df = df.drop(columns=['Total'])  #notice we have to store the value in df for this to work

df
```

## Rearranging columns 

```python
#Rearranging columns
cols = list(df.columns.values)
df = df[cols[0:4] + [cols[-1]] + cols[4:11]]

df.head(5) #'Legendary' cloumn is switched over to left side 
```

## Saving our data

```python
#Saving the data in csv without the index 
df.to_csv('modified_pokemon_dataset.csv', index=False)
```

## Filtering Data 

```python
#Filtering Data based on multiple conditions
df.loc[((df['Type 1'] =="Grass") & (df['Type 2'] == 'Poison')) | (df['Legendary'])]    #Seperate your conditions with parentheses otherwise it does not work
#In pd lib we use & instead of and, | instead of or, ~ instead of !
```

```python
#Storing Filtered Data in new csv file
new_df = df.loc[(df['Type 1'] =="Grass") & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]

new_df.to_csv('filtered_data.csv')

new_df
```

## Reset index 

```python
#Resetting Index
new_df = new_df.reset_index(drop=True) #you can drop the old index otherwise it'll also be in the data as a different column
#new_df.reset_index(drop=True, inplace=True) This is another way to store the data using inplace
new_df
```

## Regex Filtering 

```python
#Regex Filtering
import re

df.loc[df['Name'].str.contains('mega', flags=re.I, regex = True)]    #more complicated version -> [a-z]*MEGA[a-z]*.[a-z]*
#flags=re.I will ignore case senstivity as we can see here MEGA is all caps in our dataset
```

## Conditional Changes 

```python
#Conditonal Changes
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flamer'

df
```

```python
#Conditonal Changes (reverting back)
df.loc[df['Type 1'] == 'Flamer', 'Type 1'] = 'Fire'
#Can also change the Legendary to True but it will mess the whole dataset  df.loc[df['Type 1'] == 'Flamer', 'Legendary'] = True

df
```

```python
#Conditonal Changes (Manuplating columns if a certain conditoin is met)
df1 = pd.read_csv('modified_pokemon_dataset.csv')

df1.loc[df1['HP'] > 70, ['Generation','Legendary']] = 'TEST Value'

df1
```

```python
#Conditonal Changes pt2 (Manuplating columns if a certain conditoin is met)
df1.loc[df1['HP'] > 70, ['Generation','Legendary']] = ['TEST 1','TEST 2']

df1
```

## Aggregate Statistics (groupby) (Sum, Mean, Counting)

```python
# Aggregate Statistics (groupby)
df1.groupby(['Type 1']).mean(['HP','Attack','Defense','Sp. Atk','Sp. Def','Speed']).sort_values('Defense', ascending = False)  
#mean of all stats grouped by type 1 
```

```python
#Sum of all stats grouped by Type 1
df1.groupby(['Type 1']).sum(['HP','Attack','Defense','Sp. Atk','Sp. Def','Speed'])
```

```python
#Count of all types grouped by Type 1
df1.groupby(['Type 1']).count()
```

```python
#cleaner way to count
df1.groupby(['Type 1']).count()['#']
```

## Working with large amounts of data 

```python
#Working with large amounts of data (setting chunksize)
new_df = pd.DataFrame(columns=df.columns)   #new df with same columns as our main df

for df in pd.read_csv('modified_pokemon_dataset.csv', chunksize = 5):           #i.e. 5 rows are being passed in at a time
    print("CHUNK DF")
    print(df)
    results = df.groupby(['Type 1']).count()   #performing action on our data chunk
    
    new_df = pd.concat([new_df, results])   #store data in new data frame
    
```

```python
new_df   #see the data is stored with different count groups for 'Type 1' as we appended data in loops for 5 row chunks
```

# 2. Seaborn Tutorial

```python
#Seaborn is a library used for data visualization. It is a major tool used to draw insights using charts and graphs
#These visualization sare often reffered to as plots in python.
#It is built on top of matplotlib library which is th emost commonly used data visualization library in python.
import seaborn as sns
sns.set()

#Read the data
df = pd.read_csv('Pokemon.csv')
df = df.head(10)

df
```

## Scatter Plot

```python
graph = sns.scatterplot(x='Attack', y='Defense', data=df)

#To display we can also import matplotlib as plt and use the plt.show command
graph
```

## Barplot 

```python
res = sns.barplot(x='Attack', y='Defense', data=df)

res
```

```python
#we can also give specific color to any of our graphs
res = sns.barplot(x='Attack', y='Defense', data=df, color = 'yellow')  

res
```

```python
#we can also give specific color pallet to any of our graphs, you can search seaborn docs for diffrent pallets.
res = sns.barplot(x='Attack', y='Defense', data=df, palette = 'Set2')  

res
```

## Countplot 

```python
res = sns.countplot(x='Attack', data=df, palette = 'Set1')  

res
```

```python
res = sns.countplot(y='Defense', data=df, palette = 'Set2')  

res
```

```python
res = sns.countplot(x='Defense', hue ='Type 1' , data=df, palette = 'Set1')  #We can also get grouped data for these graphs plotting using hue

res
```

# 3. Bokeh Tutorial  

```python
#Bokeh is a python library for creating interactive data visualization for modern web browsers.
#You can use Python to build beautiful data visualizations, ranging from simple plots to complex dashboards with streaming datasets. 
#Bokeh lets you create HTML and JavaScript-powered visualizations without having to write any HTML or JavaScript yourself.

from bokeh.io import show, output_notebook        
# load Bokeh output in Jupyter notebook import output_ontebook
# also import show to display various graphs that we'll be plotting 
#(output_notebook is not really Required but lets you see the outputs in jupyter notebook and not an individual Tab for graph every time you hit run)

from bokeh.plotting import figure 
#Helps us in plotting all of the various graphs we'll be looking at 
# bokeh.plotting can also be used to call the show function

output_notebook()
```

## Interactive Plotting

```python
p = figure(title="Pokemon Dataset", x_axis_label='Attack', y_axis_label='Defense', width = 400, height = 400, tools="hover") 
#We have given info about what will be displayed on the graph and the shape of our scatter plot will be a square(400px*400px)
#tools is what we use to make it an interactive graph

p.circle('Attack', 'Defense', size=10, source=df1) #the first arg is x-axis, the second is y-axis, the size of the circle is 10 and the source is df1 dataset)
show(p) #display the graph (the graph would be displayed on a new tab if we didn't use output_notebook)
```

```python
# Another way of making our graph interactive
from bokeh.models import HoverTool #Used to make our cursor be able to interact with the graph

# Creating a paradigm for the info displayed by each plot on the graph
TOOLTIPS = HoverTool(tooltips=[
    ("index", "$index"),
    ("(x,y)", "($x, $y)"),
    ("(Attack,Defence)", "(@Attack, @Defense)"),
    ("Name", "@Name")]
)
# The list tooltips is made up of labels and values are supplied as a list of (label, value) tuples
# The prefixes '$' are “special fields”.These often correspond to values that are part of the plot, such as the coordinates of the mouse in data or screen space
#and '@' are associated with columns in a ColumnDataSource 
#Note that if a column name contains spaces, it must be surrounded by curly braces. -> @{column name}

p = figure(title="Pokemon Dataset", x_axis_label='Attack', y_axis_label='Defense', width=600, height=500, tools=[TOOLTIPS])
#use tools=[TOOLTIPS], Notice that the graph is more definative than the above graph
p.circle('Attack', 'Defense', size=7, source=df1)
show(p)
```
